##2.1 Expandindo os scouts

Podemos notar que o campo **Scout**, na tabela **Atletas**, contém dicionários com os scouts que o atleta acumula ao longo do campeonato. Desejamos criar uma nova coluna para cada tipo de scout possível.

```
scouts = {'G': [], 'A': [], 'FT': [], 'FD': [], 'FF': [], 'FS': [], 'PP': [], 'I': [], 'PE': [],
          'SG': [], 'DP': [], 'DD': [], 'RB': [], 'GC': [], 'CV': [], 'CA': [], 'GS': [], 'FC': []}

for row in atletas_df.iterrows(): #cada row é uma tupla (índice, série)  
    scouts_jogador = row[1]['scout']   #buscamos o elemento 1 da row (série) e, dentro da série o valor cujo índice é 'scout'
    for tipo_scout in scouts:
      if tipo_scout in scouts_jogador:
        scouts[tipo_scout].append(scouts_jogador[tipo_scout])
      else:
        scouts[tipo_scout].append(0)
```

```
scouts_df = pd.DataFrame(scouts)
scouts_df.head()
```

```
atletas_df = pd.concat([atletas_df, scouts_df], axis=1)
atletas_df.head()
```

##2.2 Incluindo informações e separando os scouts da última rodada

O DataFrame que temos até agora apresenta apenas os scouts que cada atleta acumulou ao longo de todo o campeonato. Mas não seria interessante termos os registros partida por partida? Isso é possível, desde que tenhamos um registro de como estava o mercado no rodada anterior. Felizmente, eu venho salvando estas "fotografias" rodada após rodada.

Adicionalmente, vamos incluir alguns campos que o Cartola não entrega de "mão beijada":

- **'jogou'**: **True** para quem atuou na rodada e **False** para quem não atuou;
- **'minutos'**: por padrão, os atletas que atuaram receberão valor **90** e **0** para os que não atuaram (depois eu altero estes valores manualmente);
- **'casa'**: **True** para os clubes que atuaram em casa e **False** para os que atuaram fora; e
- **'adversario_id'**: identificação do adversário.

Primeiramente precisamos descobrir qual foi a última rodada. Para isso, vamos utilizar as bibliotecas **datetime** e **dateutil**. 

```
from datetime import datetime
from dateutil import tz

# Obter a data e hora atuais de Brasília (= São Paulo)
utc = tz.gettz('UTC')
local = tz.gettz('America/Sao_Paulo')
hoje_utc = hoje = datetime.today()
hoje_utc = hoje_utc.replace(tzinfo=utc)
hoje = hoje_utc.astimezone(local) # hoje é um objeto tipo datetime contendo valores de ano, mês, dia, hora, min, seg, mseg e fuso horário

# Importar a API Rodadas do Cartola
r = requests.get('https://api.cartolafc.globo.com/rodadas')
rodadas = r.json()

# Descobrir a última e a próxima rodada
for rodada in rodadas:
  data_inicio = datetime.strptime(rodada['inicio'],'%Y-%m-%d %H:%M:%S')
  data_inicio = data_inicio.replace(tzinfo=local)
  data_fim = datetime.strptime(rodada['fim'],'%Y-%m-%d %H:%M:%S')
  data_fim = data_fim.replace(tzinfo=local)
  
  if hoje > data_fim:
    rodada_passada = rodada['rodada_id']
    proxima_rodada = rodada['rodada_id'] + 1
    
print(f'Rodada passada: {ultima_rodada} | Próxima rodada: {proxima_rodada}')
```

Próximo passo: importar a "fotografia" que tínhamos na rodada retrasada (rodada_passada -1). Para isso, precisamos fazer o upload do arquivo.

Para convencionar, eu criei a seguinte nomenclatura:
- **scouts_acumulados**: tabela que obtivemos ao final da seção 2.1 - cada atleta possui um registro com a soma de seus scouts ao longo de todo o campeonato.

- **scouts_isolados**: tabela com os scouts de uma única rodada.

- **scouts_compilados**: é a concatenação de todas as tabelas scouts_isolados existentes até o momento.

```
rodada_retrasada = rodada_passada - 1
scouts_acumulados_rodada_retrasada = pd.read_csv(f'scouts_acumulados_rodada{rodada_retrasada}.csv')
del scouts_acumulados_rodada_retrasada['Unnamed: 0'] # O índice é salvo no arquivo .csv como uma coluna
scouts_acumulados_rodada_retrasada.head()
```

Agora, vamos criar uma tabela scouts_isolados com os campos novos.

```
scouts_isolados_rodada_passada = atletas_df
scouts_isolados_rodada_passada['jogou'] = False
scouts_isolados_rodada_passada['minutos'] = 0
scouts_isolados_rodada_passada['casa'] = None
scouts_isolados_rodada_passada['adversario_id'] = None
```

O desafio agora é preencher estes novos campos, ou seja:

- Identificar os jogadores que atuaram (como titulares ou entrando no decorrer da partida);
- Descobrir se o jogador jogou em Casa ou Fora; e
- Descobrir contra qual equipe o jogador jogou.

Antes de mais nada, vamos precisar criar duas tabelas de relacionamentos: uma tendo no índice os clubes que atuaram em casa e outra tendo no índice os clubes que atuaram fora de casa.

```
r = requests.get(f'https://api.cartolafc.globo.com/partidas/{rodada_passada}')
partidas_rodada_passada_dict = r.json()
partidas_rodada_passada_dict = partidas_rodada_passada_dict['partidas']
partidas_rodada_passada = pd.DataFrame(partidas_rodada_passada_dict)

relacionamento_casa = partidas_rodada_passada[['clube_casa_id', 'clube_visitante_id']]
relacionamento_casa.columns = ['clube_id', 'adversario_id']
relacionamento_casa.set_index('clube_id', inplace=True)
relacionamento_casa
```

```
relacionamento_fora = partidas_rodada_passada[['clube_visitante_id', 'clube_casa_id']]
relacionamento_fora.columns = ['clube_id', 'adversario_id']
relacionamento_fora.set_index('clube_id', inplace=True)
relacionamento_fora
```

```
tipos_de_scouts = ['G', 'A', 'FT', 'FD', 'FF', 'FS',
                   'PP', 'I', 'PE', 'SG', 'DP', 'DD', 
                   'RB', 'GC', 'CV', 'CA', 'GS', 'FC']


for atleta in scouts_isolados_rodada_passada.iterrows():

  # 1º) Identificar atletas que atuaram
  atleta_id = atleta[1]['atleta_id']
  
  # Precisamos considerar duas hipóteses: 
  
  # H1 - Atletas que estavam no mercado na rodada retrasada
  if atleta_id in scouts_acumulados_rodada_retrasada['atleta_id'].unique():
    indice_rodada_retrasada = scouts_acumulados_rodada_retrasada.query(f'atleta_id == {atleta_id}').index[0]

    # Para saber se um atleta jogou ou não vamos comparar o número de partidas da rodada retrasada com a rodada anterior
    jogou_flag = atletas_df.loc[atleta[0], 'jogos_num'] - scouts_acumulados_rodada_retrasada.loc[indice_rodada_retrasada, 'jogos_num']

    if jogou_flag == 1:
      scouts_isolados_rodada_passada.loc[atleta[0], 'jogou'] = True
      scouts_isolados_rodada_passada.loc[atleta[0], 'minutos'] = 90

    # Para obter os scouts da rodada anterior basta subtrair os acumulados da rodada anterior pelo acumulados da rodada retrasada:    
    for tipo in tipos_de_scouts:
      acumulado_passada = atletas_df.loc[atleta[0], tipo]
      acumulado_retrasada = scouts_acumulados_rodada_retrasada.loc[indice_rodada_retrasada, tipo]

      scouts_isolados_rodada_passada.loc[atleta[0], tipo] = acumulado_passada - acumulado_retrasada
    
  # H2 - Atletas que não estavam no mercado na rodada retrasada
  else:  
    if scouts_isolados_rodada_passada.loc[atleta[0], 'jogos_num'] == 1: # Atletas que jogaram
      scouts_isolados_rodada_passada.loc[atleta[0], 'jogou'] = True
      scouts_isolados_rodada_passada.loc[atleta[0], 'minutos'] = 90
      
  # 2º) Descobrir se jogou em casa e contra qual equipe
  clube_id = atleta[1]['clube_id']
    
  if clube_id in partidas_rodada_passada['clube_casa_id'].unique():
    scouts_isolados_rodada_passada.loc[atleta[0], 'casa'] = True
    scouts_isolados_rodada_passada.loc[atleta[0], 'adversario_id'] = relacionamento_casa.loc[clube_id, 'adversario_id']
  else:
    scouts_isolados_rodada_passada.loc[atleta[0], 'casa'] = False
    scouts_isolados_rodada_passada.loc[atleta[0], 'adversario_id'] = relacionamento_fora.loc[clube_id, 'adversario_id']

scouts_isolados_rodada_passada.head()
```

```
scouts_isolados_rodada_passada.to_csv(f'scouts_isolados_rodada{rodada_passada}.csv', encoding='utf-8')
```

Salvamos o DataFrame e agora vamos concatenar os **scouts_isolados** da rodada passada com os **scouts_compilados** até a rodada retrasada.

Lembrando que é preciso fazer o upload do arquivo.

```
scouts_compilados = pd.read_csv(f'scouts_compilados_rodada{rodada_retrasada}.csv')
del scouts_compilados['Unnamed: 0']
scouts_compilados = pd.concat([scouts_compilados, scouts_isolados_rodada_passada])
scouts_compilados.index = range(scouts_compilados.shape[0])
scouts_compilados.shape
```

Maravilha! Agora é só salvar o nosso DataFrame em .csv. Vamos salvar novamente o DataFrame **atletas_df** com um novo nome para usarmos posteriormente. É **fundamental** fazer o download destes arquivos, pois eles serão apagados quando fecharmos o notebook.

```
scouts_compilados.to_csv(f'scouts_compilados_rodada{rodada_anterior}.csv', encoding='utf-8')
atletas_df.to_csv(f'scouts_acumulados_rodada{rodada_anterior}.csv', encoding='utf-8')
```
